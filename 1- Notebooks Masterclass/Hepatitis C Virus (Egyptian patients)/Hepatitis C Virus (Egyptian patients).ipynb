```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/hepatitis-c-virus-egyptian-patients/HCV-Egypt-Data.csv')

# Display basic information about the dataset
print("Basic Information about the Dataset:")
df.info()
```

```python
df.describe()
```

**What is the distribution of gender in the dataset?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each gender
gender_counts = df['Gender'].value_counts()

# Plot the distribution of gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Distribution of Gender in the Dataset')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Display gender counts
print("Gender Distribution:")
print(gender_counts)
```

**What is the correlation matrix of numerical variables in the dataset?**

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

**What is the distribution of BMI (Body Mass Index) in the dataset?**

```python
# Plot the distribution of BMI
plt.figure(figsize=(10, 6))
sns.histplot(df['BMI'], bins=20, kde=True, color='salmon')
plt.title('Distribution of BMI in the Dataset')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

**How does the RNA Base and RNA EOT (End of Treatment) levels vary in the dataset?**

```python
# Create a scatter plot for RNA Base vs RNA EOT
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RNA Base', y='RNA EOT', data=df, color='green')
plt.title('RNA Base vs RNA EOT Levels')
plt.xlabel('RNA Base')
plt.ylabel('RNA EOT')
plt.show()
```

**What is the distribution of Baseline Histological Grading and Staging?**

```python
# Create subplots for Baseline Histological Grading and Staging
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.countplot(x='Baseline histological Grading', data=df, palette='pastel')
plt.title('Distribution of Baseline Histological Grading')
plt.xlabel('Baseline Histological Grading')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.countplot(x='Baselinehistological staging', data=df, palette='muted')
plt.title('Distribution of Baseline Histological Staging')
plt.xlabel('Baseline Histological Staging')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

**How do the liver enzyme levels (AST, ALT) change over different time points (ALT 1, ALT 4, ALT 12, etc.)?**

```python
# Create subplots for AST and ALT levels over different time points
enzyme_columns = ['AST 1', 'ALT 1', 'ALT4', 'ALT 12', 'ALT 24', 'ALT 36', 'ALT 48', 'ALT after 24 w']

plt.figure(figsize=(15, 8))
for i, enzyme_column in enumerate(enzyme_columns, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(x=enzyme_column, data=df, color='purple')
    plt.title(f'{enzyme_column} Levels Over Time')
    plt.xlabel(enzyme_column)
    plt.ylabel('Enzyme Level')

plt.tight_layout()
plt.show()
```

**What is the distribution of White Blood Cell (WBC) counts in the dataset?**

```python
# Plot the distribution of WBC counts
plt.figure(figsize=(10, 6))
sns.histplot(df['WBC'], bins=20, kde=True, color='teal')
plt.title('Distribution of White Blood Cell (WBC) Counts')
plt.xlabel('WBC Counts')
plt.ylabel('Count')
plt.show()
```

**How does the baseline RNA level correlate with the RNA levels at different time points (RNA 4, RNA 12, RNA EOT, RNA EF)?**

```python
# Create a heatmap for correlation between RNA levels
rna_columns = ['RNA Base', 'RNA 4', 'RNA 12', 'RNA EOT', 'RNA EF']
rna_correlation_matrix = df[rna_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(rna_correlation_matrix, cmap='Blues', annot=True, fmt=".2f")
plt.title('Correlation Between Baseline RNA and RNA Levels at Different Time Points')
plt.show()
```

**What is the distribution of Platelet counts in the dataset?**

```python
# Plot the distribution of Platelet counts
plt.figure(figsize=(10, 6))
sns.histplot(df['Plat'], bins=20, kde=True, color='coral')
plt.title('Distribution of Platelet Counts')
plt.xlabel('Platelet Counts')
plt.ylabel('Count')
plt.show()
```

**How does the distribution of Hemoglobin (HGB) levels vary in the dataset?**

```python
# Plot the distribution of Hemoglobin levels
plt.figure(figsize=(10, 6))
sns.histplot(df['HGB'], bins=20, kde=True, color='indigo')
plt.title('Distribution of Hemoglobin (HGB) Levels')
plt.xlabel('HGB Levels')
plt.ylabel('Count')
plt.show()
```

**How do Liver Enzyme levels (AST and ALT) vary at different time points for male and female patients?**

```python
# Create subplots for AST and ALT levels over different time points, segmented by gender
plt.figure(figsize=(40, 8))
for i, enzyme_column in enumerate(enzyme_columns, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(x=enzyme_column, y='Gender', data=df, palette='viridis')
    plt.title(f'{enzyme_column} Levels Over Time by Gender')
    plt.xticks(rotation=90)
    plt.xlabel(enzyme_column)
    plt.ylabel('Gender')

plt.tight_layout()
plt.show()
```

**How do Hemoglobin (HGB) levels correlate with RNA Base levels?**

```python
# Create a scatter plot for HGB vs RNA Base levels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HGB', y='RNA Base', data=df, color='blue')
plt.title('Hemoglobin (HGB) vs RNA Base Levels')
plt.xlabel('HGB Levels')
plt.ylabel('RNA Base Levels')
plt.show()
```

**Can we visualize the relationship between Liver Enzyme levels (ALT 24) and Platelet counts?**

**How does the distribution of Hemoglobin (HGB) levels vary with increasing Baseline Histological Staging?**

```python
# Create subplots for scatter plots of HGB vs Baseline Histological Staging
plt.figure(figsize=(15, 10))
for i, staging_level in enumerate(df['Baselinehistological staging'].unique(), 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x='HGB', y='Baselinehistological staging', data=df[df['Baselinehistological staging'] == staging_level], color='orange')
    plt.title(f'HGB vs Baseline Histological Staging - Level {staging_level}')
    plt.xlabel('HGB Levels')
    plt.ylabel('Baseline Histological Staging')

plt.tight_layout()
plt.show()
```
