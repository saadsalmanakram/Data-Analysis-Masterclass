**What is the overall trend in total births in Japan from 1899 to 2022?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/kaggle/input/japan-birth-statistics/japan_birth.csv')

# Plotting the overall trend in total births
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['birth_total'], marker='o', linestyle='-', color='b')
plt.title('Overall Trend in Total Births in Japan (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Total Births')
plt.grid(True)
plt.show()
```

**How has the birth rate in Japan changed over the years?**

```python
# Plotting the birth rate trend
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['birth_rate'], marker='o', linestyle='-', color='r')
plt.title('Birth Rate Trend in Japan (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.grid(True)
plt.show()
```

**What is the distribution of infant deaths in Japan over the years?**

```python
# Plotting the distribution of infant deaths
plt.figure(figsize=(10, 6))
plt.bar(data['year'], data['infant_death_total'], color='g', label='Total Infant Deaths')
plt.bar(data['year'], data['infant_death_unknown_gender'], color='orange', label='Unknown Gender')
plt.title('Distribution of Infant Deaths in Japan (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Infant Deaths')
plt.legend()
plt.grid(True)
plt.show()
```

**How does the average age of mothers change with birth order?**

```python
# Plotting the average age of mothers by birth order
plt.figure(figsize=(10, 6))
plt.plot(data['firstborn'], data['mother_age_firstborn'], marker='o', linestyle='-', label='Firstborn')
plt.plot(data['secondborn'], data['mother_age_secondborn'], marker='o', linestyle='-', label='Secondborn')
plt.plot(data['thirdborn'], data['mother_age_thirdborn'], marker='o', linestyle='-', label='Thirdborn')
plt.title('Average Age of Mothers by Birth Order in Japan (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Average Age of Mothers')
plt.legend()
plt.grid(True)
plt.show()
```

**What is the ratio of legitimate to illegitimate children born in Japan?**

```python
# Plotting the ratio of legitimate to illegitimate children
plt.figure(figsize=(8, 6))
plt.bar(data['year'], data['legitimate_child'], color='blue', label='Legitimate')
plt.bar(data['year'], data['illegitimate_child'], color='red', label='Illegitimate')
plt.title('Ratio of Legitimate to Illegitimate Children in Japan (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Children')
plt.legend()
plt.grid(True)
plt.show()
```

**How has the total fertility rate in Japan evolved over the years?**

```python
# Plotting the total fertility rate trend
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['total_fertility_rate'], marker='o', linestyle='-', color='purple')
plt.title('Total Fertility Rate Trend in Japan (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Total Fertility Rate')
plt.grid(True)
plt.show()
```

**What is the distribution of stillbirths based on gestational weeks in Japan?**

```python
# Plotting the distribution of stillbirths based on gestational weeks
fig, ax = plt.subplots(figsize=(12, 8))
bar_width = 0.2

ax.bar(data['year'] - bar_width * 2, data['weeks_under_28'], width=bar_width, label='Under 28 weeks')
ax.bar(data['year'] - bar_width, data['weeks_28-31'], width=bar_width, label='28-31 weeks')
ax.bar(data['year'], data['weeks_32-36'], width=bar_width, label='32-36 weeks')
ax.bar(data['year'] + bar_width, data['weeks_37-41'], width=bar_width, label='37-41 weeks')
ax.bar(data['year'] + bar_width * 2, data['weeks_over_42'], width=bar_width, label='Over 42 weeks')

plt.title('Distribution of Stillbirths in Japan by Gestational Weeks (1899-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Stillbirths')
plt.legend()
plt.grid(True)
plt.show()
```
