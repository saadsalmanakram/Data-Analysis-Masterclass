**What is the distribution of respondents' gender?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("/kaggle/input/digital-habits-and-employment-status-survey/June 10-July 12 2015 - Gaming Jobs and Broadband - CSV.csv")

# Count the occurrences of each gender
gender_counts = data['sex'].value_counts()

# Plot the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='viridis')
plt.title('Distribution of Respondents\' Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**How many respondents use the internet and mobile devices?**

```python
# Count the occurrences of internet and mobile device usage
internet_usage_counts = data['eminuse'].value_counts()
mobile_usage_counts = data['intmob'].value_counts()

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=internet_usage_counts.index, y=internet_usage_counts.values, palette='muted')
plt.title('Internet Usage Distribution')
plt.xlabel('Internet Usage')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.barplot(x=mobile_usage_counts.index, y=mobile_usage_counts.values, palette='muted')
plt.title('Mobile Device Usage Distribution')
plt.xlabel('Mobile Device Usage')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
