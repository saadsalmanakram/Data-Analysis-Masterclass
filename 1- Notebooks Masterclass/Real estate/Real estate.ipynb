```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load the dataset
df = pd.read_csv('/kaggle/input/real-estate/Real estate valuation data set.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Display the number of rows and columns in the dataset
rows, columns = df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# Display summary statistics for numerical columns
summary_statistics = df.describe()
print("Summary Statistics for Numerical Columns:")
print(summary_statistics)
```

**How does the number of convenience stores vary across the dataset?**

```python
# Plot a bar chart for the 'number of convenience stores' variable
plt.figure(figsize=(10, 6))
sns.countplot(x='X4 number of convenience stores', data=df)
plt.title('Number of Convenience Stores Distribution')
plt.xlabel('Number of Convenience Stores')
plt.ylabel('Count')
plt.show()
```

**How are the transactions distributed across different years?**

```python
# Convert 'X1 transaction date' to datetime
df['X1 transaction date'] = pd.to_datetime(df['X1 transaction date'])

# Extract the year from 'X1 transaction date'
df['Transaction Year'] = df['X1 transaction date'].dt.year

# Plot a bar chart for transaction distribution across years
plt.figure(figsize=(10, 6))
sns.countplot(x='Transaction Year', data=df, palette='viridis')
plt.title('Transaction Distribution Across Years')
plt.xlabel('Transaction Year')
plt.ylabel('Count')
plt.show()
```

**How is the distribution of distances to the nearest MRT station?**

```python
# Plot a bar chart for the 'distance to the nearest MRT station' variable
plt.figure(figsize=(14, 6))
sns.histplot(df['X3 distance to the nearest MRT station'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Distance to Nearest MRT Station')
plt.xlabel('Distance to Nearest MRT Station')
plt.ylabel('Frequency')
plt.show()
```

**How is the geographical distribution of the properties based on latitude and longitude?**

```python
# Plot a scatter plot for latitude and longitude
plt.figure(figsize=(12, 8))
sns.scatterplot(x='X5 latitude', y='X6 longitude', data=df, color='green')
plt.title('Geographical Distribution of Properties')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

**How does the house price of unit area vary with the number of convenience stores?**

```python
# Plot a bar chart for the relationship between 'number of convenience stores' and 'house price of unit area'
plt.figure(figsize=(12, 6))
sns.barplot(x='X4 number of convenience stores', y='Y house price of unit area', data=df, palette='muted')
plt.title('House Price of Unit Area vs. Number of Convenience Stores')
plt.xlabel('Number of Convenience Stores')
plt.ylabel('House Price of Unit Area')
plt.show()
```
