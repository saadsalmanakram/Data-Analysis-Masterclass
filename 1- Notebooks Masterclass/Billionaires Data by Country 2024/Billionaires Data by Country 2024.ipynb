```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/billionaires-data-by-country-2024/billionaires-by-country-2024.csv')

# Display the structure of the dataset
df.info()
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Display summary statistics of numerical columns
df.describe()
```

```python
# Count the number of unique countries
unique_countries = df['country'].nunique()
print("Number of unique countries:", unique_countries)
```

**What are the top 5 countries with the highest population in 2024?**

```python
# Sort and display the top 5 countries with the highest population
top_population_countries = df.sort_values(by='population_2024', ascending=False).head(5)
print("Top 5 countries with the highest population in 2024:")
print(top_population_countries[['country', 'population_2024']])
```

**What is the average population density per square kilometer across all countries?**

```python
# Calculate and display the average population density per square kilometer
average_density = df['population_density_km'].mean()
print("Average population density per square kilometer:", average_density)
```

**How many billionaires are there per million people on average in 2023?**

```python
# Calculate and display the average number of billionaires per million people
average_billionaires_per_million = df['BillionairesPerMillionPeople2023'].mean()
print("Average billionaires per million people in 2023:", average_billionaires_per_million)
```

**What is the total net worth of the world's billionaires in 2023?**

```python
# Calculate and display the total net worth of all billionaires in 2023
total_net_worth = df['BillionairesTotalNetWorth2023'].sum()
print("Total net worth of billionaires in 2023:", total_net_worth)
```

**What is the distribution of the population growth rates across different countries?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of population growth rates
plt.figure(figsize=(10, 6))
sns.histplot(df['Population_growthRate'], bins=20, kde=True)
plt.title('Distribution of Population Growth Rates')
plt.xlabel('Population Growth Rate')
plt.ylabel('Frequency')
plt.show()
```

**How is the population density distributed across different regions?**

```python
# Plot the distribution of population density across different regions
plt.figure(figsize=(12, 8))
sns.boxplot(x='region', y='population_density_km', data=df)
plt.title('Population Density Distribution Across Regions')
plt.xlabel('Region')
plt.ylabel('Population Density per Square Kilometer')
plt.xticks(rotation=45)
plt.ylim(0,750)
plt.show()
```

**Who is the richest billionaire in 2023 and what is their net worth?**

```python
# Find the richest billionaire and their net worth in 2023
richest_billionaire = df.loc[df['BillionairesRichestNetWorth2023'].idxmax()]
print("Richest billionaire in 2023:")
print("Name:", richest_billionaire['BillionairesRichestBillionaire2023'])
print("Net Worth:", richest_billionaire['BillionairesRichestNetWorth2023'])
```

**Visualizing the Distribution of Billionaires' Net Worth in 2023**

```python
# Plotting the distribution of billionaires' net worth in 2023
plt.figure(figsize=(12, 8))
sns.histplot(df['BillionairesTotalNetWorth2023'], bins=30, kde=True)
plt.title('Distribution of Billionaires\' Net Worth in 2023')
plt.xlabel('Net Worth (in Billion Dollars)')
plt.ylabel('Frequency')
plt.show()
```

**Visualizing the Relationship between Population Size and Number of Billionaires**

```python
# Scatter plot to visualize the relationship between population size and number of billionaires
plt.figure(figsize=(20, 8))
sns.scatterplot(x='population_2024', y='BillionairesTotalNetWorth2023', data=df)
plt.title('Relationship between Population Size and Number of Billionaires in 2023')
plt.xlabel('Population Size')
plt.xticks(rotation=90)
plt.ylabel('Number of Billionaires')
plt.show()
```

**Visualizing the Proportion of Billionaires per Million People in 2023 by Country**

```python
# Bar plot to visualize the proportion of billionaires per million people in 2023 by country
plt.figure(figsize=(16, 8))
sns.barplot(x='country', y='BillionairesPerMillionPeople2023', data=df, palette='viridis')
plt.title('Proportion of Billionaires per Million People in 2023 by Country')
plt.xlabel('Country')
plt.ylabel('Billionaires per Million People')
plt.xticks(rotation=90)
plt.show()
```

**Visualizing the Population Density and Growth Rate by Country**

```python
# Scatter plot to visualize the population density and growth rate by country
plt.figure(figsize=(14, 8))
sns.scatterplot(x='population_density_km', y='Population_growthRate', data=df, hue='country', palette='Set2', legend=False)
plt.title('Population Density vs. Growth Rate by Country')
plt.xlabel('Population Density per Square Kilometer')
plt.ylabel('Population Growth Rate')
plt.show()
```

**Visualizing the Top 10 Countries with the Highest Billionaires per Million People in 2023**

```python
# Bar plot to visualize the top 10 countries with the highest billionaires per million people in 2023
top_countries_billionaires_per_million = df.nlargest(10, 'BillionairesPerMillionPeople2023')

plt.figure(figsize=(12, 8))
sns.barplot(x='BillionairesPerMillionPeople2023', y='country', data=top_countries_billionaires_per_million, palette='mako')
plt.title('Top 10 Countries with the Highest Billionaires per Million People in 2023')
plt.xlabel('Billionaires per Million People')
plt.ylabel('Country')
plt.show()
```

**Visualizing the Relationship between Population Density and the Number of Billionaires**
