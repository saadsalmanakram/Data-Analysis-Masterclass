```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load each dataset
cases_of_infectious_diseases = pd.read_csv('/kaggle/input/pandemics-in-world/1- the-number-of-cases-of-infectious-diseases.csv')
vaccinated_one_year_olds = pd.read_csv('/kaggle/input/pandemics-in-world/2- the-worlds-number-of-vaccinated-one-year-olds.csv')
annual_influenza_mortality = pd.read_csv('/kaggle/input/pandemics-in-world/3- annual-mortality-rate-from-seasonal-influenza-ages-65.csv')
excess_deaths_cumulative = pd.read_csv('/kaggle/input/pandemics-in-world/4- excess-deaths-cumulative-economist-single-entity.csv')
cholera_deaths = pd.read_csv('/kaggle/input/pandemics-in-world/5- number-of-reported-cholera-deaths.csv')

# Display basic information about each dataset
print("Cases of Infectious Diseases:")
print(cases_of_infectious_diseases.info())

print("\nVaccinated One-Year-Olds:")
print(vaccinated_one_year_olds.info())

print("\nAnnual Influenza Mortality:")
print(annual_influenza_mortality.info())

print("\nExcess Deaths Cumulative:")
print(excess_deaths_cumulative.info())

print("\nCholera Deaths:")
print(cholera_deaths.info())
```

**Clean the data to handle missing values, duplicates, and outliers.**

```python
# Handling missing values
cases_of_infectious_diseases.dropna(inplace=True)
vaccinated_one_year_olds.dropna(inplace=True)
annual_influenza_mortality.dropna(inplace=True)
excess_deaths_cumulative.dropna(inplace=True)
cholera_deaths.dropna(inplace=True)

# Handling duplicates
cases_of_infectious_diseases.drop_duplicates(inplace=True)
vaccinated_one_year_olds.drop_duplicates(inplace=True)
annual_influenza_mortality.drop_duplicates(inplace=True)
excess_deaths_cumulative.drop_duplicates(inplace=True)
cholera_deaths.drop_duplicates(inplace=True)

# Display cleaned datasets
print("Cleaned Cases of Infectious Diseases:")
print(cases_of_infectious_diseases.head())

print("\nCleaned Vaccinated One-Year-Olds:")
print(vaccinated_one_year_olds.head())

print("\nCleaned Annual Influenza Mortality:")
print(annual_influenza_mortality.head())

print("\nCleaned Excess Deaths Cumulative:")
print(excess_deaths_cumulative.head())

print("\nCleaned Cholera Deaths:")
print(cholera_deaths.head())
```

**Investigate the vaccination coverage for one-year-olds and identify any patterns or changes over time.**

```python
# Visualization of vaccination coverage for one-year-olds
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Number of one-year-olds vaccinated with HepB3', data=vaccinated_one_year_olds, label='HepB3')
sns.lineplot(x='Year', y='Number of one-year-olds vaccinated with DTP containing vaccine, 3rd dose', data=vaccinated_one_year_olds, label='DTP 3rd Dose')
sns.lineplot(x='Year', y='Number of one-year-olds vaccinated with polio, 3rd dose', data=vaccinated_one_year_olds, label='Polio 3rd Dose')
plt.title('Vaccination Coverage for One-Year-Olds Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Vaccinated One-Year-Olds')
plt.legend()
plt.show()
```

**Examine the cumulative estimated daily excess deaths and reported cholera deaths, highlighting the impact of COVID-19.**

```python
# Visualization of cumulative estimated daily excess deaths and reported cholera deaths
plt.figure(figsize=(12, 8))
sns.lineplot(x='Day', y='cumulative_estimated_daily_excess_deaths', data=excess_deaths_cumulative, label='Excess Deaths')
sns.lineplot(x='Year', y='Reported cholera deaths', data=cholera_deaths, label='Cholera Deaths')
plt.title('Cumulative Estimated Daily Excess Deaths vs. Reported Cholera Deaths')
plt.xlabel('Time')
plt.ylabel('Number of Deaths')
plt.legend()
plt.show()
```

**Explore the trends in the number of reported cases of malaria, HIV/AIDS, and tuberculosis over the years.**

**Explore the relationship between the number of one-year-olds vaccinated with rotavirus (last dose) and the number of reported cholera cases.**

```python
# Relationship analysis between the number of one-year-olds vaccinated with rotavirus (last dose) and reported cholera cases
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number of one-year-olds vaccinated with rotavirus, last dose', y='Reported cholera cases', data=vaccinated_one_year_olds.merge(cases_of_infectious_diseases, on=['Entity', 'Code', 'Year']))
plt.title('Relationship between Rotavirus Vaccination and Reported Cholera Cases')
plt.xlabel('Rotavirus Vaccination (Last Dose)')
plt.ylabel('Reported Cholera Cases')
plt.show()
```

**Investigate the seasonality or trends in the annual mortality rate from seasonal influenza for individuals aged 65 and older without using line plots.**

```python
# Seasonality or trend analysis in influenza mortality rate (ages 65 and older) without using line plots
plt.figure(figsize=(12, 8))
sns.stripplot(x='Year', y='rate over65', data=annual_influenza_mortality, jitter=True)
plt.title('Seasonality or Trends in Influenza Mortality Rate (Ages 65 and Older) Over Time')
plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.show()
```

**Create a Plotly Sunburst chart to visualize the distribution of reported cholera deaths across different entities.**

```python
import plotly.express as px

# Sunburst chart using Plotly for distribution of reported cholera deaths
fig = px.sunburst(cholera_deaths, path=['Entity', 'Year'], values='Reported cholera deaths', 
                  title='Distribution of Reported Cholera Deaths Across Entities and Years')
fig.show()
```

**Use Plotly to create a choropleth map displaying the reported cholera deaths across different entities.**

```python
import plotly.express as px

# Choropleth map using Plotly for reported cholera deaths across different entities
fig = px.choropleth(cholera_deaths, 
                    locations='Entity', 
                    locationmode='country names',
                    color='Reported cholera deaths',
                    hover_name='Entity',
                    title='Reported Cholera Deaths Across Different Entities')
fig.update_geos(showcoastlines=True, coastlinecolor="Black", showland=True, landcolor="LightGray")
fig.show()
```
