```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Loading the dataset
df = pd.read_csv('/kaggle/input/operating-system-market-2009-2023/os_worldwide.csv')

# Displaying the first few rows of the dataset
df.head()
```

```python
df.info()
```

**Are there any missing values in the dataset? If yes, how many in each column?**

```python
# Checking for missing values in each column
missing_values = df.isnull().sum()

# Displaying the number of missing values in each column
missing_values
```

**How is the statistical summary of the dataset?**

```python
# Displaying the statistical summary of the dataset
summary_stats = df.describe()

# Transposing the summary for better readability
summary_stats.T
```

**How is the distribution of the 'Windows' market share over the years?**

```python
# Plotting the distribution of 'Windows' market share
plt.figure(figsize=(10, 6))
sns.histplot(df['Windows'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Windows Market Share')
plt.xlabel('Windows Market Share (%)')
plt.ylabel('Frequency')
plt.show()
```

**How has the market share of 'Android' changed over the years?**

```python
# Plotting the trend of 'Android' market share over time
plt.figure(figsize=(30, 8))
sns.lineplot(x='Date', y='Android', data=df)
plt.title('Trend of Android Market Share Over the Years')
plt.xlabel('Date')
plt.ylabel('Android Market Share (%)')
plt.xticks(rotation=90)
plt.show()
```

**Can we identify and handle any outliers in the dataset?**

```python
# Checking for outliers using box plots for each operating system
plt.figure(figsize=(16, 8))
sns.boxplot(data=df.drop(columns=['Date']))
plt.title('Box Plots for Operating System Market Share')
plt.xlabel('Operating Systems')
plt.ylabel('Market Share (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Are there any duplicate rows in the dataset, and if so, how many?**

```python
# Checking for duplicate rows in the dataset
duplicate_rows = df[df.duplicated()]
num_duplicate_rows = len(duplicate_rows)

# Displaying the number of duplicate rows
num_duplicate_rows
```

**How has the overall market share evolved over the years?**

```python
# Calculating the total market share for each month
df['Total_Market_Share'] = df.drop(columns=['Date']).sum(axis=1)

# Plotting the trend of overall market share over time
plt.figure(figsize=(30, 8))
sns.lineplot(x='Date', y='Total_Market_Share', data=df)
plt.title('Trend of Overall Market Share Over the Years')
plt.xlabel('Date')
plt.ylabel('Total Market Share (%)')
plt.xticks(rotation=90)
plt.show()
```

**What is the distribution of market shares for the 'iOS' operating system?**

```python
# Plotting the distribution of 'iOS' market share
plt.figure(figsize=(10, 6))
sns.histplot(df['iOS'], bins=20, kde=True, color='salmon')
plt.title('Distribution of iOS Market Share')
plt.xlabel('iOS Market Share (%)')
plt.ylabel('Frequency')
plt.show()
```

**How has the market share of 'Chrome OS' changed over the years?**

```python
# Plotting the trend of 'Chrome OS' market share over time
plt.figure(figsize=(30, 8))
sns.lineplot(x='Date', y='Chrome OS', data=df)
plt.title('Trend of Chrome OS Market Share Over the Years')
plt.xlabel('Date')
plt.ylabel('Chrome OS Market Share (%)')
plt.xticks(rotation=90)
plt.show()
```

**What is the distribution of market shares for the 'Linux' operating system?**

```python
# Plotting the distribution of 'Linux' market share
plt.figure(figsize=(10, 6))
sns.histplot(df['Linux'], bins=20, kde=True, color='green')
plt.title('Distribution of Linux Market Share')
plt.xlabel('Linux Market Share (%)')
plt.ylabel('Frequency')
plt.show()
```

**Are there any months with extremely low overall market share? If so, how many?**

```python
# Identifying months with total market share less than 1%
low_market_share_months = df[df['Total_Market_Share'] < 1]
num_low_market_share_months = len(low_market_share_months)

# Displaying the number of months with low total market share
num_low_market_share_months
```

**How has the market share of 'iOS' compared to 'Android' over the years?**

```python
# Plotting the trend of 'iOS' and 'Android' market share over time
plt.figure(figsize=(30, 8))
sns.lineplot(x='Date', y='iOS', data=df, label='iOS')
sns.lineplot(x='Date', y='Android', data=df, label='Android')
plt.title('Trend of iOS and Android Market Share Over the Years')
plt.xlabel('Date')
plt.ylabel('Market Share (%)')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

**What is the average market share for each operating system?**

```python
# Calculating the average market share for each operating system
average_market_share = df.drop(columns=['Date', 'Total_Market_Share']).mean()

# Displaying the average market share for each operating system
average_market_share
```

**Are there any noticeable trends or patterns in the market share data?**

```python
# Plotting the trend of market share for each operating system over time
plt.figure(figsize=(30, 10))
for os in df.columns[1:]:
    sns.lineplot(x='Date', y=os, data=df, label=os)

plt.title('Trend of Market Share for Each Operating System Over the Years')
plt.xlabel('Date')
plt.ylabel('Market Share (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

**How has the market share of 'iOS' changed over time using Plotly?**

```python
import plotly.express as px

# Creating a line plot for the trend of 'iOS' market share using Plotly
fig = px.line(df, x='Date', y='iOS', title='Trend of iOS Market Share Over Time',
              labels={'iOS': 'iOS Market Share (%)', 'Date': 'Date'})
fig.update_xaxes(type='category')
fig.show()
```

**Can we create an interactive bar chart for the average market share of each operating system using Plotly?**

```python
# Calculating the average market share for each operating system
average_market_share_plotly = pd.DataFrame({'Operating System': average_market_share.index,
                                            'Average Market Share': average_market_share.values})

# Creating an interactive bar chart using Plotly
fig = px.bar(average_market_share_plotly, x='Operating System', y='Average Market Share',
             title='Average Market Share for Each Operating System',
             labels={'Average Market Share': 'Average Market Share (%)', 'Operating System': 'Operating System'})
fig.show()
```

**How has the market share of the top 5 operating systems evolved over the years?**

```python
# Selecting the top 5 operating systems based on average market share
top_5_os = average_market_share.nlargest(5).index

# Plotting the trend of market share for the top 5 operating systems over time
plt.figure(figsize=(30, 10))
for os in top_5_os:
    sns.lineplot(x='Date', y=os, data=df, label=os)

plt.title('Trend of Market Share for Top 5 Operating Systems Over the Years')
plt.xlabel('Date')
plt.ylabel('Market Share (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

**What is the distribution of market shares for the top 3 operating systems?**

```python
# Selecting the top 3 operating systems based on average market share
top_3_os = average_market_share.nlargest(3).index

# Creating a box plot for the distribution of market shares for the top 3 operating systems
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[top_3_os])
plt.title('Box Plot for Market Share Distribution of Top 3 Operating Systems')
plt.xlabel('Operating Systems')
plt.ylabel('Market Share (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**How has the market share of 'Android' compared to 'iOS' using a bar chart?**

```python
# Creating a bar chart for the comparison of 'Android' and 'iOS' market share
plt.figure(figsize=(10, 6))
sns.barplot(x=['Android', 'iOS'], y=[average_market_share['Android'], average_market_share['iOS']], palette='viridis')
plt.title('Comparison of Android and iOS Average Market Share')
plt.xlabel('Operating System')
plt.ylabel('Average Market Share (%)')
plt.show()
```

**Can we create a sunburst chart to visualize the market share hierarchy?**

```python
import plotly.express as px

# Creating a sunburst chart to visualize market share hierarchy
fig = px.sunburst(df, path=['Date'] + df.columns[1:].tolist(), values=df['Total_Market_Share'],
                  title='Market Share Hierarchy Over Time')
fig.update_layout(margin=dict(l=0, r=0, b=0, t=40))
fig.show()
```

**How has the market share of the top 3 operating systems changed over time using an animated line chart?**

```python
# Creating an animated line chart for the top 3 operating systems
fig = px.line(df, x='Date', y=top_3_os, title='Animated Trend of Top 3 Operating Systems Over Time',
              labels={'value': 'Market Share (%)', 'variable': 'Operating System'})
fig.update_layout(xaxis=dict(type='category'), yaxis=dict(range=[0, 35]))  # Adjust the y-axis range
fig.show()
```

**What is the distribution of market shares for all operating systems in a 3D scatter plot?**

```python
# Creating a 3D scatter plot for the distribution of market shares
fig = px.scatter_3d(df.melt(id_vars=['Date']), x='Date', y='variable', z='value',
                    color='value', size='value', size_max=30, opacity=0.7,
                    title='3D Scatter Plot of Market Shares for All Operating Systems')
fig.update_layout(scene=dict(zaxis=dict(range=[0, 30])))  # Adjust the z-axis range
fig.show()
```
