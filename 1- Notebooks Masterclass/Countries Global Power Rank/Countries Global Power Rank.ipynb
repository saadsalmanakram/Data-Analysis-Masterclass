**What is the distribution of the Military Strength Rankings in 2023?**

```python
import plotly.express as px
import pandas as pd

# Load the dataset
df = pd.read_csv("/kaggle/input/countries-global-power-rank/2023 Military Strength Ranking3.csv")
```

**What is the distribution of the "Rank" variable in the dataset?**

```python
# Plot the distribution of Rank
fig = px.histogram(df, x="Rank", nbins=20, title="Distribution of Military Strength Rank")
fig.show()
```

**What is the correlation between "Total Aircraft" and "Fighter Aircrafts"?**

```python
# Plot the correlation between Total Aircraft and Fighter Aircrafts
fig = px.scatter(df, x="Total Aircraft", y="Fighter Aircrafts", title="Correlation between Total Aircraft and Fighter Aircrafts")
fig.show()
```

**What is the relationship between "Active Mil Personnel" and "Defense Budget"?**

```python
# Plot the relationship between Active Mil Personnel and Defense Budget
fig = px.scatter(df, x="Active Mil Personnel", y="Defense Budget", title="Relationship between Active Mil Personnel and Defense Budget")
fig.show()
```

**How do the numbers of "Army Personnel," "Navy Personnel," and "Air Force Personnel" compare across countries?**

```python
# Plot the comparison of Army, Navy, and Air Force Personnel
fig = px.bar(df, x="Rank", y=["Army Personnel", "Navy Personnel", "Air Force Personnel"], 
             title="Comparison of Army, Navy, and Air Force Personnel Across Countries")
fig.show()
```
