```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/nyc-languages/nyc-language-data.csv')

# Display basic information about the dataset
df.info()
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Count the number of unique languages
unique_languages = df['Language'].nunique()
print(f'The dataset contains {unique_languages} unique languages.')
```

```python
# Display the unique world regions
world_regions = df['World Region'].unique()
print('World Regions:', world_regions)
```

```python
# Display the unique language families
language_families = df['Language Family'].unique()
print('Language Families:', language_families)
```

```python
# Display the distribution of language statuses
status_distribution = df['Status'].value_counts()
print('Language Status Distribution:\n', status_distribution)
```

```python
# Count the number of languages under each status category
status_counts = df['Status'].value_counts()
print('Number of Languages in Each Status Category:\n', status_counts)
```

**Can we visualize the distribution of language sizes?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Plot the distribution of language sizes
sns.histplot(df['Size'], bins=20, kde=True)

# Set plot labels and title
plt.xlabel('Language Size')
plt.ylabel('Frequency')
plt.title('Distribution of Language Sizes')

# Show the plot
plt.show()
```

**What is the distribution of languages across different world regions?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Plot the distribution of languages across world regions using a bar plot
sns.countplot(x='World Region', data=df, palette='viridis')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set plot labels and title
plt.xlabel('World Region')
plt.ylabel('Number of Languages')
plt.title('Distribution of Languages Across World Regions')

# Show the plot
plt.show()
```

**What is the distribution of language families in the dataset?**

```python
# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Plot the distribution of language families using a bar plot
sns.countplot(x='Language Family', data=df, palette='muted', order=df['Language Family'].value_counts().index)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set plot labels and title
plt.xlabel('Language Family')
plt.ylabel('Number of Languages')
plt.title('Distribution of Language Families')

# Show the plot
plt.show()
```

**How many languages fall under each status category?**

```python
# Set up the figure and axis
plt.figure(figsize=(8, 5))

# Plot the count of languages in each status category using a bar plot
sns.countplot(x='Status', data=df, palette='pastel')

# Set plot labels and title
plt.xlabel('Language Status')
plt.ylabel('Number of Languages')
plt.title('Number of Languages in Each Status Category')

# Show the plot
plt.show()
```

**What is the distribution of languages in each size category?**

```python
# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Plot the distribution of languages in each size category using a bar plot
sns.countplot(x='Size', data=df, palette='viridis')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set plot labels and title
plt.xlabel('Language Size')
plt.ylabel('Number of Languages')
plt.title('Distribution of Languages in Each Size Category')

# Show the plot
plt.show()
```

**What is the relationship between the number of global speakers and language size on a logarithmic scale?**

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Scatter plot with logarithmic scale for both axes
sns.scatterplot(x='Global Speakers', y='Size', data=df, color='skyblue')

# Set logarithmic scale for both axes
plt.xscale('log')
plt.yscale('log')

# Set plot labels and title
plt.xlabel('Global Speakers (log scale)')
plt.ylabel('Language Size (log scale)')
plt.title('Relationship between Global Speakers and Language Size (log scale)')

# Show the plot
plt.show()
```

**How does the distribution of global speakers vary across different language families on a logarithmic scale?**

```python
# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Scatter plot with logarithmic scale for y-axis
sns.scatterplot(x='Language Family', y='Global Speakers', data=df)

# Set logarithmic scale for the y-axis
plt.yscale('log')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set plot labels and title
plt.xlabel('Language Family')
plt.ylabel('Global Speakers (log scale)')
plt.title('Distribution of Global Speakers Across Language Families (log scale)')

# Show the plot
plt.show()
```

**Explore the relationship between language size and the number of global speakers based on their status on a logarithmic scale.**

```python
# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Scatter plot with logarithmic scale for both axes, color-coded by status
sns.scatterplot(x='Size', y='Global Speakers', hue='Status', data=df, palette='viridis', s=80)

# Set logarithmic scale for both axes
plt.xscale('log')
plt.yscale('log')

# Set plot labels and title
plt.xlabel('Language Size (log scale)')
plt.ylabel('Global Speakers (log scale)')
plt.title('Relationship between Language Size, Global Speakers, and Status (log scale)')

# Show the plot
plt.show()
```

**Explore the relationship between language size and the number of global speakers based on their status, with a limit of 10 on both axes.**

```python
# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Scatter plot with a limit of 10 on both axes, color-coded by status
sns.scatterplot(x='Size', y='Global Speakers', hue='Status', data=df, palette='viridis', s=80)

# Set limits for both axes
plt.xlim(1, 10)
plt.ylim(1, 10)

# Set plot labels and title
plt.xlabel('Language Size (Up to 10)')
plt.ylabel('Global Speakers (Up to 10)')
plt.title('Relationship between Language Size, Global Speakers, and Status (Limited to 10)')

# Show the plot
plt.show()
```

**How does the distribution of global speakers vary across different language families with a limit of 10 on the y-axis?**

**Explore the relationship between the number of global speakers and language size on a logarithmic scale, with a limit of 10 on both axes.**

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Scatter plot with logarithmic scale for both axes
sns.scatterplot(x='Global Speakers', y='Size', data=df, color='skyblue')

# Set logarithmic scale for both axes
plt.xscale('log')
plt.yscale('log')

# Set limits for both axes
plt.xlim(1, 10)
plt.ylim(1, 10)

# Set plot labels and title
plt.xlabel('Global Speakers (log scale, Up to 10)')
plt.ylabel('Language Size (log scale, Up to 10)')
plt.title('Relationship between Global Speakers and Language Size (log scale, Limited to 10)')

# Show the plot
plt.show()
```

**Explore the distribution of languages across different language families using a pie chart.**

```python
# Calculate the distribution of languages across language families
language_family_distribution = df['Language Family'].value_counts()

# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Plot a pie chart for the distribution of languages across language families
plt.pie(language_family_distribution, labels=language_family_distribution.index, autopct='%1.1f%%', colors=sns.color_palette('Set3'))

# Set plot title
plt.title('Distribution of Languages Across Language Families')

# Show the plot
plt.show()
```

**What is the distribution of language statuses using a pie chart?**

```python
# Calculate the distribution of language statuses
status_distribution = df['Status'].value_counts()

# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Plot a pie chart for the distribution of language statuses
plt.pie(status_distribution, labels=status_distribution.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))

# Set plot title
plt.title('Distribution of Language Statuses')

# Show the plot
plt.show()
```
