# **3. Data Cleaning**

**3.1 Identifying Outliers using Z-Score**

```python
from scipy import stats

# Calculate Z-scores of each value in the DataFrame
z_scores_df1 = stats.zscore(df1.select_dtypes(include=['float64', 'int64']))
z_scores_df2 = stats.zscore(df2.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
df1_outliers = df1[(z_scores_df1 > 3).any(axis=1) | (z_scores_df1 < -3).any(axis=1)]
df2_outliers = df2[(z_scores_df2 > 3).any(axis=1) | (z_scores_df2 < -3).any(axis=1)]
```

**3.2 Remove Outliers**

```python
# Remove rows with outliers based on Z-score
df1_no_outliers = df1[(z_scores_df1 < 3).all(axis=1) & (z_scores_df1 > -3).all(axis=1)]
df2_no_outliers = df2[(z_scores_df2 < 3).all(axis=1) & (z_scores_df2 > -3).all(axis=1)]
```

**Let's visualize the distribution of the grades.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Distribution of G1, G2, and G3
plt.figure(figsize=(12, 6))
sns.histplot(df1['G1'], kde=True, color='blue', label='G1')
sns.histplot(df1['G2'], kde=True, color='green', label='G2')
sns.histplot(df1['G3'], kde=True, color='red', label='G3')
plt.legend()
plt.title("Distribution of Grades (G1, G2, G3) - File1")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()
```

**We'll create a heatmap to visualize the correlation between numerical features.**

```python
# Assuming df1 is your DataFrame
# Filter numeric columns
numeric_df = df1.select_dtypes(include=[float, int])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(14, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap - File1")
plt.show()
```

**We'll create a heatmap to show how average final grades vary by school and sex.**

```python
import numpy as np

# Calculate the average final grade (G3) by school and sex
pivot_table = df1.pivot_table(values='G3', index='school', columns='sex', aggfunc=np.mean)

plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.1f', linewidths=0.5)
plt.title('Average Final Grade (G3) by School and Sex - File1')
plt.xlabel('Sex')
plt.ylabel('School')
plt.show()
```

**We can compare the distribution of final grades (G3) between the two datasets.**

**We can compare the average study time between the two datasets.**

```python
# Calculate the average study time by school
studytime_pivot_df1 = df1.pivot_table(values='studytime', index='school', aggfunc=np.mean)
studytime_pivot_df2 = df2.pivot_table(values='studytime', index='school', aggfunc=np.mean)

comparison_df = pd.DataFrame({
    'File1': studytime_pivot_df1['studytime'],
    'File2': studytime_pivot_df2['studytime']
})

comparison_df.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Average Study Time by School')
plt.xlabel('School')
plt.ylabel('Average Study Time (hours)')
plt.show()
```
