```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the dataset
df = pd.read_csv('/kaggle/input/price-of-used-toyota-corolla-cars/ToyotaCorolla.csv')

# Display the first few rows of the dataset
df.head()
```

```python
df.info()
```

**How to Handle Missing Values?**

```python
# Check for missing values in the dataset
print(df.isnull().sum())

# Drop rows with missing values or use imputation methods if necessary
# Example: Drop rows with any missing values
df = df.dropna()

# Verify that missing values are handled
print(df.isnull().sum())
```

**What are the Summary Statistics of Numeric Columns?**

```python
df.describe()
```

**What is the Distribution of the Target Variable 'Price'?**

```python
# Explore the distribution of the target variable 'Price'
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

**What is the Correlation Between Numeric Variables?**

```python
# Extract numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate and visualize the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()

```

**How to Visualize the Age Distribution of Cars**

```python
import plotly.express as px

# Visualize the age distribution of cars using Plotly
fig = px.histogram(df, x='Age_08_04', nbins=30, title='Age Distribution of Cars',
                   labels={'Age_08_04': 'Age (months)', 'count': 'Count'})
fig.update_layout(xaxis_title='Age (months)', yaxis_title='Count')
fig.show()
```

**What is the Relationship Between Engine Size and Horsepower with a 3D Scatter Plot?**

```python
# Visualize the relationship between engine size, horsepower, and price using Plotly
fig = px.scatter_3d(df, x='CC', y='HP', z='Price', title='Engine Size, Horsepower, and Price',
                    labels={'CC': 'Engine Size (CC)', 'HP': 'Horsepower', 'Price': 'Price'})
fig.show()
```
