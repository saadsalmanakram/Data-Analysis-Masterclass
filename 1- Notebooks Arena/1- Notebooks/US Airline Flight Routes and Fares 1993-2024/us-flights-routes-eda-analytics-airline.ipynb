# **3. Data Cleaning**

**3.1 Identifying Outliers using Z-Score**

```python
from scipy import stats

# Calculate Z-scores of each value in the DataFrame
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
df_outliers = df[(z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)]
print(df_outliers.shape)
```

**3.2 Remove Outliers**

```python
# Remove rows with outliers based on Z-score
df_no_outliers = df[(z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)]
df_no_outliers.shape
```

# **Let's the flight Begin**

**How are the fare and nsmiles distributed?**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Distribution of fare
plt.figure(figsize=(10, 6))
sns.histplot(df['fare'], kde=True, color='blue')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# Distribution of nsmiles
plt.figure(figsize=(10, 6))
sns.histplot(df['nsmiles'], kde=True, color='green')
plt.title('Distribution of Miles')
plt.xlabel('Miles')
plt.ylabel('Frequency')
plt.show()
```

**What is the relationship between fare and nsmiles?**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='nsmiles', y='fare', data=df)
plt.title('Relationship between Fare and Miles')
plt.xlabel('Miles')
plt.ylabel('Fare')
plt.show()
```

**What is the average fare by year?**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='fare', data=df.groupby('Year')['fare'].mean().reset_index(), marker='o')
plt.title('Average Fare by Year')
plt.xlabel('Year')
plt.ylabel('Average Fare')
plt.show()
```

**How does the passenger count vary across different city1?**

```python
plt.figure(figsize=(12, 8))
top_cities = df.groupby('city1')['passengers'].sum().nlargest(10).reset_index()
sns.barplot(x='passengers', y='city1', data=top_cities)
plt.title('Top 10 Cities by Passenger Count')
plt.xlabel('Total Passengers')
plt.ylabel('City 1')
plt.show()
```

**What is the distribution of carrier_lg and carrier_low market shares?**

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='carrier_lg', y='large_ms', data=df)
plt.title('Distribution of Large Carrier Market Share')
plt.xlabel('Carrier')
plt.xticks(rotation=90)
plt.ylabel('Market Share')
plt.show()

plt.figure(figsize=(15, 6))
sns.boxplot(x='carrier_low', y='lf_ms', data=df)
plt.title('Distribution of Low Carrier Market Share')
plt.xlabel('Carrier')
plt.xticks(rotation=90)
plt.ylabel('Market Share')
plt.show()
```

**What are the most common airport_1 and airport_2 pairs?**

```python
airport_pairs = df.groupby(['airport_1', 'airport_2']).size().nlargest(10).reset_index(name='count')
plt.figure(figsize=(12, 8))
sns.barplot(x='count', y='airport_1', hue='airport_2', data=airport_pairs)
plt.title('Top 10 Airport Pairs')
plt.xlabel('Number of Routes')
plt.ylabel('Airport 1')
# Move the legend to the right
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

**Which city1 and city2 combinations have the highest average fare?**

```python
# Calculate the average fare for each city pair
city_pairs_fare = df.groupby(['city1', 'city2'])['fare'].mean().reset_index()

# Sort by the highest average fare and take the top 10
top_city_pairs_fare = city_pairs_fare.nlargest(10, 'fare')

plt.figure(figsize=(12, 8))
sns.barplot(x='fare', y='city1', hue='city2', data=top_city_pairs_fare)
plt.title('Top 10 City Pairs by Average Fare')
plt.xlabel('Average Fare')
plt.ylabel('City 1')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

**How does the passenger count vary by quarter and year?**

```python
plt.figure(figsize=(20, 8))
sns.boxplot(x='quarter', y='passengers', hue='Year', data=df)
plt.title('Passenger Count Variation by Quarter and Year')
plt.xlabel('Quarter')
plt.ylabel('Passenger Count')
plt.legend(title='Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

**What are the busiest airports in terms of passenger count?**

```python
# Group by airport_1 and airport_2 separately and sum passengers
airport1_passengers = df.groupby('airport_1')['passengers'].sum().reset_index()
airport2_passengers = df.groupby('airport_2')['passengers'].sum().reset_index()

# Concatenate the results
busiest_airports = pd.concat([airport1_passengers, airport2_passengers], ignore_index=True)
busiest_airports = busiest_airports.groupby('airport_1')['passengers'].sum().reset_index()

# Top 10 busiest airports
top_airports = busiest_airports.nlargest(10, 'passengers')

plt.figure(figsize=(12, 8))
sns.barplot(x='passengers', y='airport_1', data=top_airports)
plt.title('Top 10 Busiest Airports by Passenger Count')
plt.xlabel('Total Passengers')
plt.ylabel('Airport')
plt.show()
```

**How do fare and passenger counts vary across different carrier_lg?**

```python
plt.figure(figsize=(20, 8))
sns.boxplot(x='carrier_lg', y='fare', data=df)
plt.title('Fare Distribution Across Large Carriers')
plt.xlabel('Carrier')
plt.ylabel('Fare')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(20, 8))
sns.boxplot(x='carrier_lg', y='passengers', data=df)
plt.title('Passenger Count Distribution Across Large Carriers')
plt.xlabel('Carrier')
plt.ylabel('Passengers')
plt.xticks(rotation=90)
plt.show()
```

**Which city1 has the most routes originating from it?**

```python
# Count the number of unique city2 for each city1
city_routes = df.groupby('city1')['city2'].nunique().reset_index().sort_values(by='city2', ascending=False)

# Top 10 cities with the most routes
top_city_routes = city_routes.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='city2', y='city1', data=top_city_routes)
plt.title('Top 10 Cities by Number of Routes Originating')
plt.xlabel('Number of Routes')
plt.ylabel('City 1')
plt.show()
```

**How do fare_lg and fare_low compare for the same routes?**

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='fare_lg', y='fare_low', data=df)
plt.title('Comparison of Large and Low Carrier Fares for Same Routes')
plt.xlabel('Fare of Large Carrier')
plt.ylabel('Fare of Low Carrier')
plt.show()
```

**How does market share (large_ms vs lf_ms) vary across different year?**

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='large_ms', data=df, label='Large Carrier Market Share')
sns.lineplot(x='Year', y='lf_ms', data=df, label='Low Carrier Market Share', linestyle='--')
plt.title('Market Share of Large vs Low Carriers Over the Years')
plt.xlabel('Year')
plt.ylabel('Market Share')
plt.legend()
plt.show()
```
