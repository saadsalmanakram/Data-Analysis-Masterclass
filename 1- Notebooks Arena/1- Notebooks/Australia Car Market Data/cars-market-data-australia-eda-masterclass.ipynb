# **3- Data Cleaning**

**3.1- Identifying Outliers using Z-Score:**

```python
from scipy import stats

# Calculate Z-scores of each value in the DataFrame
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
df_outliers = df[(z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)]
print(f"Number of rows with outliers: {len(df_outliers)}")
```

**3.2- Remove Outliers:**

```python
# Remove rows with outliers based on Z-score
df_no_outliers = df[(z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)]
print(f"Shape of data after removing outliers: {df_no_outliers.shape}")
```

# **4. Analysis**

**Visualizing the distribution of 'Price' and 'Kilometers'**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Visualizing the distribution of 'Price' and 'Kilometers'
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Price'], kde=True, bins=30, color='blue')
plt.title('Distribution of Price')

plt.subplot(1, 2, 2)
sns.histplot(df['Kilometers'], kde=True, bins=30, color='green')
plt.title('Distribution of Kilometers')

plt.show()
```

**Bar plot showing the count of cars by Brand**

```python
import plotly.express as px

# Bar plot showing the count of cars by Brand
brand_counts = df['Brand'].value_counts()
fig = px.bar(brand_counts, x=brand_counts.index, y=brand_counts.values, 
             labels={'x': 'Brand', 'y': 'Count'}, title="Car Brand Popularity")
fig.show()
```

**Price Comparison by Fuel Type**

```python
# Box plot showing Price distribution across different Fuel types
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel', y='Price', data=df)
plt.title('Price Comparison by Fuel Type')
plt.xticks(rotation=90)
plt.show()
```

**Car Age vs. Price Scatter Plot**

```python
# Calculating Car Age
df['Car_Age'] = 2024 - df['Year']

# Scatter plot of Car Age vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Car_Age', y='Price', data=df)
plt.title('Car Age vs. Price')
plt.show()
```

**Correlation Matrix**

```python
import numpy as np

# Correlation matrix of numerical features
plt.figure(figsize=(12, 8))

# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
