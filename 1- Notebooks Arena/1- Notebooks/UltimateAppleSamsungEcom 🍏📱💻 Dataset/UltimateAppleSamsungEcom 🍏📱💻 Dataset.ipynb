**Visualize the distribution of iPhone conditions on Amazon using a bar plot.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each condition in iPhone Amazon data
iphone_amazon_condition_counts = iphone_amazon_df['condition'].value_counts()

# Plot a bar chart for iPhone Amazon conditions
plt.figure(figsize=(8, 6))
sns.barplot(x=iphone_amazon_condition_counts.index, y=iphone_amazon_condition_counts.values, palette="viridis")
plt.title('Distribution of iPhone Conditions on Amazon')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.show()
```

**Can you visualize the distribution of prices for iPhones on Amazon from the 'iphone_amazon.csv' file?**

**Explore the relationship between 'price' and 'review_count' for both iPhone and Samsung phones on Amazon using scatter plots.**

```python
# Concatenate iPhone and Samsung Amazon datasets
combined_amazon_df = pd.concat([iphone_amazon_df, samsung_amazon_df], keys=['iPhone', 'Samsung'])

# Scatter plot to explore the relationship between 'price' and 'review_count'
plt.figure(figsize=(12, 8))
sns.scatterplot(x='price', y='review_count', hue=combined_amazon_df.index.get_level_values(0), data=combined_amazon_df)
plt.title('Relationship between Price and Review Count for iPhone and Samsung Phones on Amazon')
plt.xlabel('Price')
plt.xticks(rotation=90)
plt.ylabel('Review Count')
plt.legend(title='Brand')
plt.show()
```

**Visualize the distribution of ratings for both iPhone and Samsung phones on Amazon based on different conditions using scatter plots.**

```python
# Scatter plot to visualize the distribution of ratings based on different conditions for iPhone and Samsung phones on Amazon
combined_amazon_df['brand'] = combined_amazon_df.index.get_level_values(0)
plt.figure(figsize=(12, 8))
sns.scatterplot(x='condition', y='rating', hue='brand', data=combined_amazon_df)
plt.title('Distribution of Ratings for iPhone and Samsung Phones on Amazon by Condition')
plt.xlabel('Condition')
plt.ylabel('Rating')
plt.legend(title='Brand')
plt.show()
```

**Create a bar plot to compare the number of products in each condition for iPhones and Samsung phones on Amazon and eBay.**

```python
# Concatenate the condition columns for iPhones and Samsung phones on Amazon and eBay
combined_conditions = pd.concat([iphone_amazon_df['condition'], iphone_ebay_df['condition'],
                                 samsung_amazon_df['condition'], samsung_ebay_df['condition']],
                                keys=['iPhone_Amazon', 'iPhone_eBay', 'Samsung_Amazon', 'Samsung_eBay'])

# Bar plot to compare the number of products in each condition for iPhones and Samsung phones on Amazon and eBay
plt.figure(figsize=(12, 8))
sns.countplot(x=combined_conditions.values, hue=combined_conditions.index.get_level_values(0))
plt.title('Comparison of Product Conditions for iPhones and Samsung Phones on Amazon and eBay')
plt.xlabel('Condition')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.legend(title='Brand and Platform')
plt.show()
```
