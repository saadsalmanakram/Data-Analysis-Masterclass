```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/lucknow-city-india-450-housing-prices-updated/lucknow_hosuing_prices - Main File.csv')

# Display basic statistics
basic_stats = df.describe()
print(basic_stats)
```

```python
# Count unique locations and display them
unique_locations = df['location'].nunique()
locations_list = df['location'].unique()

print(f"Number of unique locations: {unique_locations}")
print("List of unique locations:", locations_list)
```

**What is the distribution of the number of bedrooms (BHK) in the dataset?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of BHK
plt.figure(figsize=(8, 6))
sns.countplot(x='bhk', data=df)
plt.title('Distribution of Number of Bedrooms (BHK)')
plt.xlabel('Number of Bedrooms (BHK)')
plt.ylabel('Count')
plt.show()
```

**How is the relationship between the area in square feet and the housing prices (in lakh)?**

```python
# Scatter plot for area_sq_ft vs. price_lakh
plt.figure(figsize=(10, 6))
sns.scatterplot(x='area_sq_ft', y='price_lakh', data=df)
plt.title('Relationship between Area and Housing Prices')
plt.xlabel('Area (Square Feet)')
plt.ylabel('Price (Lakh)')
plt.show()
```

```python
# Calculate average price for each type
average_price_by_type = df.groupby('type')['price_lakh'].mean().sort_values(ascending=False)
print("Average Price by Housing Type:")
print(average_price_by_type)
```

```python
# Count of houses in each status category
status_count = df['status'].value_counts()
print("Number of houses in each status category:")
print(status_count)
```

**Can you visualize the distribution of housing prices in Lucknow?**

```python
# Distribution plot for housing prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price_lakh'], bins=20, kde=True)
plt.title('Distribution of Housing Prices in Lucknow')
plt.xlabel('Price (Lakh)')
plt.ylabel('Frequency')
plt.show()
```

**What is the average price (in lakh) for each location, and can you visualize this information through a bar plot?**

```python
# Calculate average price for each location
average_price_by_location = df.groupby('location')['price_lakh'].mean().sort_values(ascending=False)

# Bar plot for average price by location
plt.figure(figsize=(20, 8))
average_price_by_location.plot(kind='bar', color='skyblue')
plt.title('Average Housing Prices by Location in Lucknow')
plt.xlabel('Location')
plt.ylabel('Average Price (Lakh)')
plt.xticks(rotation=90, ha='right')
plt.show()
```
