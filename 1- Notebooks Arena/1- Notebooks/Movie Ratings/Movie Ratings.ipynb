**What is the distribution of movie types (Type 1 and Type 2) in the dataset?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/kaggle/input/movie-ratings/movieRatings.csv')

# Plot the distribution of Type 1 and Type 2 movies
plt.figure(figsize=(12, 6))
df['Type 1'].value_counts().plot(kind='bar', color='skyblue', label='Type 1')
df['Type 2'].value_counts().plot(kind='bar', color='orange', label='Type 2', alpha=0.7)
plt.title('Distribution of Movie Types')
plt.xlabel('Movie Type')
plt.ylabel('Count')
plt.legend()
plt.show()
```

**How is the distribution of movie ratings among friends?**

```python
# Plot the distribution of movie ratings among friends
friend_columns = ['Friend 2 score', 'Friend 5 score', 'Friend 7 score', 'Friend 8 score']
plt.figure(figsize=(12, 6))
df[friend_columns].plot(kind='box', vert=False, patch_artist=True)
plt.title('Distribution of Movie Ratings Among Friends')
plt.xlabel('Movie Ratings')
plt.ylabel('Friends')
plt.show()
```

**What is the correlation between the duration of movies and the IMDB scores?**

```python
# Plot the correlation between movie duration and IMDB scores
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration (minutes)'], df['IMDB score'], alpha=0.5)
plt.title('Correlation Between Movie Duration and IMDB Scores')
plt.xlabel('Duration (minutes)')
plt.ylabel('IMDB Score')
plt.show()
```

**How many times did each friend fall asleep during the movies?**

```python
# Plot the number of times each friend fell asleep during movies
plt.figure(figsize=(12, 6))
df['Did Friend 5 sleep'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Number of Times Friends Fell Asleep During Movies')
plt.xlabel('Did Friend 5 Sleep?')
plt.ylabel('Count')
plt.show()
```

**What is the distribution of movie origins by country?**

```python
# Plot the distribution of movie origins by country
plt.figure(figsize=(14, 6))
df['Origin Country'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribution of Movie Origins by Country')
plt.xlabel('Origin Country')
plt.ylabel('Count')
plt.show()
```

**How many movies of each type were watched in each year?**

```python
# Plot the number of movies of each type watched in each year
plt.figure(figsize=(14, 6))
df.groupby(['Year', 'Type 1'])['ID'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Number of Movies of Each Type Watched in Each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

**What is the distribution of animation movies?**

```python
# Plot the distribution of animation movies
plt.figure(figsize=(10, 6))
df['Animation'].value_counts().plot(kind='bar', color='teal')
plt.title('Distribution of Animation Movies')
plt.xlabel('Animation')
plt.ylabel('Count')
plt.show()
```

**How many movies were chosen by each friend?**

```python
# Plot the number of movies chosen by each friend
plt.figure(figsize=(12, 6))
df['Friend Who chose it'].value_counts().plot(kind='bar', color='gold')
plt.title('Number of Movies Chosen by Each Friend')
plt.xlabel('Friend Who Chose It')
plt.ylabel('Count')
plt.show()
```

**What is the distribution of IMDB-Mean scores?**

**How many times did each friend fall asleep based on movie type?**

```python
# Plot the number of times each friend fell asleep based on movie type
plt.figure(figsize=(14, 6))
df.groupby(['Did Friend 5 sleep', 'Type 1'])['ID'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Number of Times Friends Fell Asleep Based on Movie Type')
plt.xlabel('Did Friend 5 Sleep?')
plt.legend(bbox_to_anchor=(1,1))
plt.ylabel('Count')
plt.show()
```

**What is the distribution of friend scores for movies of each type?**

```python
# Plot the distribution of friend scores for movies of each type
friend_score_columns = ['Friend 2 score', 'Friend 5 score', 'Friend 7 score', 'Friend 8 score']
plt.figure(figsize=(16, 8))
df.groupby('Type 1')[friend_score_columns].mean().plot(kind='bar')
plt.title('Distribution of Friend Scores for Movies of Each Type')
plt.xlabel('Movie Type')
plt.ylabel('Average Score')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

**How does the average movie duration vary for each friend?**

**Create a pair plot to visualize relationships between numeric variables.**

```python
import seaborn as sns

# Select relevant numeric columns for pair plot
numeric_columns = ['Duration (minutes)', 'Friend 2 score', 'Friend 5 score', 'Friend 7 score', 'Friend 8 score',
                    'Mean score', 'IMDB score', 'IMDB-Mean']

# Create pair plot
sns.pairplot(df[numeric_columns])
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```

**Create a pair plot to visualize relationships between scores given by different friends.**

```python
# Select columns related to friend scores for pair plot
friend_score_columns = ['Friend 2 score', 'Friend 5 score', 'Friend 7 score', 'Friend 8 score']

# Create pair plot for friend scores
sns.pairplot(df[friend_score_columns])
plt.suptitle('Pair Plot of Friend Scores', y=1.02)
plt.show()
```

**Create a pair plot to visualize relationships between duration, IMDB score, and mean score.**

```python
# Select columns for pair plot including duration, IMDB score, and mean score
selected_columns = ['Duration (minutes)', 'IMDB score', 'Mean score']

# Create pair plot for selected columns
sns.pairplot(df[selected_columns])
plt.suptitle('Pair Plot of Duration, IMDB Score, and Mean Score', y=1.02)
plt.show()
```

**Create a pair plot to visualize relationships between movie types and friend scores.**

```python
# Select columns for pair plot including movie types and friend scores
selected_columns = ['Type 1', 'Friend 2 score', 'Friend 5 score', 'Friend 7 score', 'Friend 8 score']

# Create pair plot for selected columns
sns.pairplot(df[selected_columns], hue='Type 1')
plt.suptitle('Pair Plot of Movie Types and Friend Scores', y=1.02)
plt.show()
```

**Create a pair plot to visualize relationships between sleep patterns and movie ratings.**

```python
# Select columns for pair plot including sleep patterns and movie ratings
selected_columns = ['Did Friend 5 sleep', 'Friend 2 score', 'Friend 5 score', 'Friend 7 score', 'Friend 8 score', 'Mean score']

# Create pair plot for selected columns
sns.pairplot(df[selected_columns], hue='Did Friend 5 sleep')
plt.suptitle('Pair Plot of Sleep Patterns and Movie Ratings', y=1.02)
plt.show()
```
