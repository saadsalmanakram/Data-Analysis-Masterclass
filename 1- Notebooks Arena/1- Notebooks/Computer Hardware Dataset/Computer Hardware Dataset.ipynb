```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import plotly.express as px

# Load datasets
cpu_cooler_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/CPUCoolerData.csv')
cpu_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/CPUData.csv')
case_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/CaseData.csv')
gpu_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/GPUData.csv')
hdd_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/HDDData.csv')
monitor_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/MonitorData.csv')
motherboard_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/MotherboardData.csv')
psu_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/PSUData.csv')
ram_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/RAMData.csv')
ssd_data = pd.read_csv('/kaggle/input/general-computer-hardware-dataset/SSDData.csv')
```

**What does the distribution of CPU cooler prices look like?**

```python
# EDA for CPU Cooler Data
fig = px.histogram(cpu_cooler_data, x='Price', title='Distribution of CPU Cooler Prices')
fig.show()
```

**How many CPUs have an integrated GPU, and what is the distribution of their base clocks?**

```python
# EDA for CPU Data
integrated_gpu_counts = cpu_data['Integrated GPU'].value_counts()
fig = px.bar(integrated_gpu_counts, x=integrated_gpu_counts.index, y=integrated_gpu_counts.values,
             labels={'index': 'Integrated GPU', 'y': 'Count'}, title='Distribution of CPUs with Integrated GPU')
fig.show()

# Distribution of Base Clocks
fig = px.histogram(cpu_data, x='Base Clock', title='Distribution of CPU Base Clocks')
fig.show()
```
