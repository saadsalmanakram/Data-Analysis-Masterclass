```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/microsoft-stock-price/microsoft_stock.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Display the number of rows and columns in the dataset
rows, columns = df.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

```python
# Display the range of dates in the dataset
date_range = df['Date'].min(), df['Date'].max()
print(f"The dataset covers the date range from {date_range[0]} to {date_range[1]}.")
```

```python
# Display summary statistics for numerical columns
summary_stats = df.describe()
print(summary_stats)
```

**How is the distribution of the 'Close' prices in the dataset?**

```python
import matplotlib.pyplot as plt

# Plot the distribution of 'Close' prices
plt.figure(figsize=(10, 6))
plt.hist(df['Close'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

**How has the 'Close' price of Microsoft stock changed over time?**

```python
import matplotlib.pyplot as plt

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Plot the 'Close' prices over time
plt.figure(figsize=(14, 8))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.title('Microsoft Stock Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

**How does the 'Open' price vary over time?**

```python
import matplotlib.pyplot as plt

# Plot the 'Open' prices over time using a bar plot
plt.figure(figsize=(14, 8))
plt.bar(df['Date'], df['Open'], color='orange', edgecolor='black')
plt.title('Microsoft Stock Open Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.show()
```

**What is the distribution of 'Adj Close' prices?**

**How does the trading volume vary over time?**

```python
# Plot the trading volume over time using a bar plot
plt.figure(figsize=(14, 8))
plt.bar(df['Date'], df['Volume'], color='brown', edgecolor='blue')
plt.title('Microsoft Stock Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

**How does the 'Close' price compare to the 'Open' price on a daily basis?**

```python
# Calculate the daily price change (Close - Open)
df['Daily_Price_Change'] = df['Close'] - df['Open']

# Plot the daily price change using a bar plot
plt.figure(figsize=(14, 8))
plt.bar(df['Date'], df['Daily_Price_Change'], color='red' if df['Daily_Price_Change'].gt(0).all() else 'green', edgecolor='green')
plt.title('Daily Price Change (Close - Open)')
plt.xlabel('Date')
plt.ylabel('Price Change')
plt.show()
```

**What is the relationship between 'High' and 'Low' prices on a daily basis?**

```python
# Calculate the daily price range (High - Low)
df['Daily_Price_Range'] = df['High'] - df['Low']

# Plot the daily price range using a bar plot
plt.figure(figsize=(14, 8))
plt.bar(df['Date'], df['Daily_Price_Range'], color='grey', edgecolor='cyan')
plt.title('Daily Price Range (High - Low)')
plt.xlabel('Date')
plt.ylabel('Price Range')
plt.show()
```

**How does the percentage change in 'Close' price compare to the previous day?**

```python
# Calculate the percentage change in 'Close' price
df['Close_Percentage_Change'] = df['Close'].pct_change() * 100

# Plot the percentage change in 'Close' price using a bar plot
plt.figure(figsize=(14, 8))
plt.bar(df['Date'], df['Close_Percentage_Change'], color='blue' if df['Close_Percentage_Change'].gt(0).all() else 'red', edgecolor='yellow')
plt.title('Percentage Change in Close Price (vs. Previous Day)')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.show()
```

**What is the distribution of trading volumes in different ranges?**

```python
# Create bins for trading volume ranges
volume_bins = [0, 5e7, 1e8, 1.5e8, 2e8, 2.5e8]
volume_labels = ['< 50M', '50M - 100M', '100M - 150M', '150M - 200M', '> 200M']

# Categorize trading volumes into bins
df['Volume_Category'] = pd.cut(df['Volume'], bins=volume_bins, labels=volume_labels, right=False)

# Create a pie chart of trading volume distribution
volume_distribution = df['Volume_Category'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(volume_distribution, labels=volume_distribution.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral', 'lightgreen', 'lightyellow', 'lightpink'])
plt.title('Distribution of Trading Volumes')
plt.show()
```

**How does the proportion of 'Up' and 'Down' daily price changes compare?**

```python
# Classify daily price changes as 'Up' or 'Down'
df['Price_Direction'] = df['Daily_Price_Change'].apply(lambda x: 'Up' if x > 0 else 'Down' if x < 0 else 'No Change')

# Create a pie chart of daily price change distribution
price_direction_distribution = df['Price_Direction'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(price_direction_distribution, labels=price_direction_distribution.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral', 'lightgrey'])
plt.title('Distribution of Daily Price Changes')
plt.show()
```

**What is the distribution of 'Close' price categories based on quartiles?**

```python
# Create quartiles for 'Close' prices
close_quartiles = pd.qcut(df['Close'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Create a pie chart of 'Close' price quartile distribution
close_quartile_distribution = close_quartiles.value_counts()
plt.figure(figsize=(10, 8))
plt.pie(close_quartile_distribution, labels=close_quartile_distribution.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow'])
plt.title('Distribution of Close Price Quartiles')
plt.show()
```
